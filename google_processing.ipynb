## Comparing different London boroughs

```python
london.drop(['country_region_code', 'country_region', 'sub_region_1',
             'metro_area', 'iso_3166_2_code', 'census_fips_code', 'place_id'], axis=1)
london
```

```python
## Looking at difference between boroughs
boroughs = london.groupby('sub_region_2').mean()
```

```python
# Removing "Borough of" from the start of every Borough name
borough_names = []
for borough in boroughs.index:
    i = borough.find('of')
    if i>0:
        borough_names.append(borough[i+3:])
    else:
        borough_names.append(borough)
borough_names[0] = 'City of London'
borough_names
```

```python
boroughs.index=borough_names
boroughs.to_csv('../datasets/google_boroughs.csv')
```
