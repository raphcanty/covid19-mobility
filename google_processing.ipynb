## Google COVID-19 Community Mobility Report data

```python
import pandas as pd
import numpy as np
```

```python
google_2020 = pd.read_csv("data_sources/2020_GB_Region_Mobility_Report.csv")
google_2020
```

```python
google_2021 = pd.read_csv("data_sources/2021_GB_Region_Mobility_Report.csv")
google_2021
```

```python
london_2020 = google_2020[google_2020.sub_region_1=='Greater London']
london_2021 = google_2021[google_2021.sub_region_1=='Greater London']
```

```python
greater_london = pd.concat([london_2020[london_2020.sub_region_2.astype('str')=='nan'],
                            london_2021[london_2021.sub_region_2.astype('str')=='nan']]) 
greater_london
```

```python
greater_london.drop(['country_region_code', 'country_region', 'sub_region_1', 'sub_region_2',
                     'metro_area', 'iso_3166_2_code', 'census_fips_code', 'place_id'], axis=1, inplace=True)
greater_london.set_index('date', inplace=True)
```

```python
greater_london.to_csv('datasets/google_greater_london.csv')
```

## Comparing different London boroughs

```python
# Looking at difference between boroughs
london_boroughs = pd.concat([london_2020, london_2021]).drop(['country_region_code', 'country_region', 'sub_region_1',
                                                              'metro_area', 'iso_3166_2_code', 'census_fips_code',
                                                              'place_id'], axis=1).set_index('date')

boroughs_l1 = london_boroughs.loc[((london_boroughs.index>'2020-03-23')&(london_boroughs.index<'2020-03-32')|
                                   (london_boroughs.index>'2020-04-00')&(london_boroughs.index<'2020-04-31')|
                                   (london_boroughs.index>'2020-05-00')&(london_boroughs.index<'2020-05-14'))].groupby('sub_region_2').mean()
```

```python
# Removing "Borough of" from the start of every Borough name
borough_names = []
for borough in boroughs_l1.index:
    i = borough.find('of')
    if i>0:
        borough_names.append(borough[i+3:])
    else:
        borough_names.append(borough)
borough_names[0] = 'City of London'
print(borough_names)
```

```python
boroughs_l1.index = borough_names
boroughs_l1.to_csv('datasets/google_boroughs.csv')
```
