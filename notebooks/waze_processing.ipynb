## Waze Processing

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
waze_old = pd.read_csv('../prelim_data/data_sources/Waze | COVID-19 Impact Dashboard_City-Level Data_Table.csv', index_col=0)
#waze.drop(['City', 'Country'], axis=1, inplace=True)
waze_old
```

```python
waze = pd.read_csv('../data_sources/Waze_City-Level_Data.csv')
waze['Date'] = pd.to_datetime(waze['Date'].apply(lambda x: datetime.strftime(datetime.strptime(x, '%b %d, %Y'), '%Y-%m-%d')))
waze
```

```python
london = waze.loc[waze.City=='London']
london = london.sort_values('Date').set_index('Date').drop(['City', 'Country'], axis=1)
london
```

```python
london.to_csv('../datasets/waze_london.csv')
```

```python
new_york = waze.loc[waze.City=='New York']
new_york = new_york.sort_values('Date').set_index('Date').drop(['City', 'Country'], axis=1)
new_york
```

```python
new_york.to_csv('../datasets/waze_new_york.csv')
```

```python
sydney = waze.loc[waze.City=='Sydney']
sydney = sydney.sort_values('Date').set_index('Date').drop(['City', 'Country'], axis=1)
sydney
```

```python
sydney.to_csv('../datasets/waze_sydney.csv')
```
