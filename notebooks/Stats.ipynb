## Statistics regarding Eat out to help out

```python
# Groups data by day of week and selects relevent time periods
grr = data['google_retail_recreation'].copy().reset_index()
grr['date'] = pd.to_datetime(grr.date)
eat_out = grr[np.isin(grr.date.dt.weekday, [0, 1, 2]) & ('2020-07-31' < grr.date) & ('2020-09-01' > grr.date)]
non_eatout = grr[np.isin(grr.date.dt.weekday, [3, 4, 5, 6]) & ('2020-07-31' < grr.date) & ('2020-09-01' > grr.date)]
sep_mtw = grr[np.isin(grr.date.dt.weekday, [0, 1, 2]) & ('2020-08-31' < grr.date) & ('2020-10-01' > grr.date)]
sep_tfss = grr[np.isin(grr.date.dt.weekday, [3, 4, 5, 6]) & ('2020-08-31' < grr.date) & ('2020-10-01' > grr.date)]
eat_out.mean(), non_eatout.mean(), sep_mtw.mean(), sep_tfss.mean()
```

```python
# Finds dates which are high for the retail & recreation category
grr[(grr.date>'2020-03-15')][grr.google_retail_recreation>75]
```

## Statistics regarding cycle hire

```python
# Reads relevent data and generates unclassified and weekend categories
cycle = pd.read_csv('../datasets/TfL_cycle.csv', usecols=[0,2,3,4,5], index_col=0, parse_dates=True)
cycle['unclassified'] = cycle['total'] - cycle[['city', 'suburb', 'city_suburb']].sum(axis=1)
cycle['weekend'] = cycle.index.weekday >= 5
cycle = cycle.loc[:'2021-07-31']
cycle
```

```python
# Finds percentage metrics from before and after lockdown
cycle_2019 = cycle.loc[:'2020-03-22']
gb_2019 = cycle_2019.groupby('weekend').sum()
(gb_2019.divide(gb_2019.total, axis=0) * 100).round(1)
```

```python
cycle_postld = cycle.loc['2020-03-23':]
gb_postld = cycle_postld.groupby('weekend').sum()
(gb_postld.divide(gb_postld.total, axis=0) * 100).round(1)
```

```python
# Finds numeric metrics from before and after lockdown
cm = cycle_2019.groupby('weekend').mean()#[['city','city_suburb','suburb','unclassified']]
cm.round()
```

```python
# Checks absolute difference between weekend and weekday
cm.loc[False] - cm.loc[True]
```

```python
cycle_postld.groupby('weekend').mean().round()#[['city','city_suburb','suburb','unclassified']]
```

```python
# Checks popular dates during lockdown 1
first_lockdown = cycle.loc['2020-03-23':'2020-05-13']
first_lockdown[first_lockdown.total>26000]
```
